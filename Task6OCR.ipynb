```python
# Перевод файла из формата rgba в формат rgb
def rgba2rgb(image_file):
    image = cv2.imread(image_file, cv2.IMREAD_UNCHANGED)
    image_mask = image[:, :, 3] == 0
    image[image_mask] = [255, 255, 255, 255]
    image_new = cv2.cvtColor(image, cv2.COLOR_BGRA2BGR)
    cv2.imwrite(image_file, image_new)
```

```python
# Изменение формата файлов в директориях
def changing_files(dir_path):
    dir = [x[0] for x in sorted(os.walk(dir_path))]
    fls = [x[2] for x in sorted(os.walk(dir_path))]
    for i in range(1, len(dir)):
        list_files = fls[i]
        for f in list_files:
            image_file = dir[i] + '/' + f
            rgba2rgb(image_file)
```

## Обучение модели распознавания букв

```python
imagedatagenerator = ImageDataGenerator

TRAINING_DIR = '/content/train'

train_datagen = imagedatagenerator(rotation_range = 15,
                                   width_shift_range = 0.1,
                                   height_shift_range = 0.1,
                                   rescale = 1.0 / 255.
                                   )

train_generator = train_datagen.flow_from_directory(TRAINING_DIR,
                                                    color_mode='grayscale',
                                                    batch_size=8,
                                                    class_mode='categorical',
                                                    target_size=(64,64)
                                                    )

VALIDATION_DIR = '/content/val'

validation_datagen = imagedatagenerator(rotation_range = 15,
                                        width_shift_range = 0.1,
                                        height_shift_range = 0.1,
                                        rescale = 1.0 / 255.
                                        )

validation_generator = validation_datagen.flow_from_directory(VALIDATION_DIR,
                                                              color_mode='grayscale',
                                                              batch_size=8,
                                                              class_mode='categorical',
                                                              target_size=(64,64)
                                                              )
```

```python
model = Sequential()
model.add(BatchNormalization(input_shape=(64,64, 1)))
model.add(Conv2D(32, (3, 3), padding='same', activation='relu'))
model.add(Conv2D(32, (3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(96, (3, 3), padding='same', activation='relu'))
model.add(Conv2D(96, (3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dense(33, activation='softmax'))
```

```python
model.compile(optimizer=Adam(learning_rate=1e-3), 
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

## Картинки для распознавания

selection_recognition_area(img_path): -> img

```python
# Для распознавания выделяется участок скана на котором находится ФИО
# высота - 36    ширина - 26
# сверху - 19    слева  - 11
# снизу  - 26    справа - 23
# сверху - 19/36 слева  - 11/26
# снизу  - 25/36 справа - 23/26
```

```python
# Выделение области распознавания
def selection_recognition_area(img_path):

    img = img_to_array(load_img(img_path))
    img_height = img.shape[0]
    img_width  = round(img_height*26/36)
    img_up     = round(img_height*19/36)
    img_down   = round(img_height*25/36)
    img_left   = round(img_width *11/26)
    img_right  = round(img_width *23/26)
    img = img[img_up : img_down, img_left : img_right]
    img = np.array(img)

    return img
```

## Определение участков с текстом

get_words_from_image(img): -> words

```python
# Выделение слов из рисунка
def get_words_from_image(img):
    words = []
    img = cv2.resize(img, (640,320))
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY).astype("uint8")
    ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    img_erode = cv2.erode(thresh, np.ones((7, 7), np.uint8), iterations=1)

    # Определение контуров
    contours, hierarchy = cv2.findContours(img_erode, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)

    # Выделение и вырезание слов по контурам
    for idx, contour in enumerate(contours):
        (x, y, w, h) = cv2.boundingRect(contour)
        if hierarchy[0][idx][3] == 0:
            word_crop = gray[y:y + h, x:x + w]
            words.append([w, h, word_crop])

    words.reverse()

    # Изменение размера слов до 64 по высоте и пропорционально по ширине
    for i in range(len(words)):
        w, h = words[i][0], words[i][1]
        words[i][0] = round(64 * w / h)
        words[i][1] = 64
        words[i][2] = cv2.resize(words[i][2], (words[i][0], words[i][1]))
    
    return words
```

## Определение контуров букв

get_letters_from_word(word): -> letters

additional_processing_letters(letters): -> letters_new

```python
def get_letters_from_word(word):
    letters = []
    img = word[2]
    ret, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    img_erode = cv2.dilate(thresh, np.ones((3, 3), np.uint8), iterations=1)

    # Определение контуров
    contours, hierarchy = cv2.findContours(img_erode, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)

    # Выделение и вырезание букв по контурам
    for idx, contour in enumerate(contours):
        (x, y, w, h) = cv2.boundingRect(contour)
    
        if hierarchy[0][idx][3] == 0:
            dh = (64 - h) // 3 + 1
            y1 = y - dh
            if y1 < 0: y1 = 0
            y2 = y + h + dh
            if y2 > word[1]: y2 = word[1]
            x1 = x - dh
            if x1 < 0: x1 = 0
            x2 = x + w + dh
            if x2 > word[0]: x2 = word[0]
        
            letter_crop = img[y1:y2, x1:x2]
        
            # Сохранение в список списка из координаты буквы,
            # её фактического размера по ширине и высоте и массива буквы
            letters.append([x1, x2 - x1, y2 - y1, letter_crop])
    
    return letters
```

```python
# Дополнительная проверка букв
def additional_processing_letters(letters):
    
    # Определение минимального соотношения ширины и высоты
    whratio = 100
    for _, w, h, _ in letters:
        whr = w / h
        if whr < whratio:
            whratio = whr

    # Разделение 2-х или 3-х букв, если они остались объединены
    letters_new = []
    for x, w, h, letter in letters:

        w2h_ratio = w / h
    
        if round(w2h_ratio / whratio) == 1:
            letters_new.append([x, w, h, letter])

        if round(w2h_ratio / whratio) == 2:
            w_new = w // 2
            x1, w1, h1, letter1 = x,         w_new, h, letter[0:h, 0:w_new]
            x2, w2, h2, letter2 = x + w_new, w_new, h, letter[0:h, w_new+1:w]
            letters_new.append([x1, w1, h1, letter1])
            letters_new.append([x2, w2, h2, letter2])

        if round(w2h_ratio / whratio) == 3:
            w_new = w // 3
            x1, w1, h1, letter1 = x,             w_new, h, letter[0:h, 0:w_new]
            x2, w2, h2, letter2 = x + w_new+1,   w_new, h, letter[0:h, w_new+1:2*w_new]
            x3, w3, h3, letter3 = x + 2*w_new+1, w_new, h, letter[0:h, 2*w_new+1:w]
            letters_new.append([x1, w1, h1, letter1])
            letters_new.append([x2, w2, h2, letter2])
            letters_new.append([x3, w3, h3, letter3])
    
    return letters_new
```

## Распознавание букв моделью нейронки

letters2word(letters): -> word

```python
# Составление слова из списка распознанных букв
def letters2word(letters):
    
    word_list = []
    for i in range(len(letters)):
        img_arr = letters[i][3]
        ret, thresh = cv2.threshold(img_arr, 127, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
        img_arr = cv2.dilate(thresh, np.ones((7, 7), np.uint8), iterations=1)
        img_arr = img_arr / 255.0
        img_arr = img_arr.reshape((1, 64, 64, 1))
        predict = model.predict(img_arr, verbose=0)
        result = np.argmax(predict)
        word_list.append(classes[result])

    word = ''.join(word_list)
    
    return word
```
