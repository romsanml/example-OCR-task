# 3) обучить собственный алгоритм, которые распознает ФИО с паспорта

```python
!pip install easyocr

import os.path
import easyocr
from PIL import Image
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# высота - 36    ширина - 26
# сверху - 19    слева  - 11
# снизу  - 26    справа - 23
# сверху - 19/36 слева  - 11/26
# снизу  - 25/36 справа - 23/26
```

```python
img_path = '/content/drive/My Drive/Projects/MindSet OCR/DataForOCR/'
list_files = sorted(os.listdir(img_path))
for filename in list_files:
    img = np.asarray(Image.open(img_path+filename))
    img_height = img.shape[0]
    img_width  = int(round(img_height*26/36, 0))
    img_up     = int(round(img_height*19/36, 0))
    img_down   = int(round(img_height*25/36, 0))
    img_left   = int(round(img_width *11/26, 0))
    img_right  = int(round(img_width *23/26, 0))
    img = Image.fromarray(img[img_up:img_down, img_left:img_right])
    img.save(filename)
```

```python
for f in list_files:
    text = easyocr.Reader(["ru"],
                          gpu=False,
                          verbose=False).readtext(f,
                                                  detail=0,
                                                  paragraph=True,
                                                  text_threshold=0.8)
    print('Файл', f, '\n', '----------', '\n', text, '\n', '----------')
```
