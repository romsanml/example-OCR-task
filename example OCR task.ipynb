## Картинки для распознавания

```python
import os
import numpy as np
from PIL import Image
```

```python
# высота - 36    ширина - 26
# сверху - 19    слева  - 11
# снизу  - 26    справа - 23
# сверху - 19/36 слева  - 11/26
# снизу  - 25/36 справа - 23/26
```

```python
# Разметка участка для распознавания
img_path = '/content/drive/My Drive/Projects/MindSet OCR/DataForOCR/'
list_files = sorted(os.listdir(img_path))
for filename in list_files:
    img = np.asarray(Image.open(img_path+filename))
    img_height = img.shape[0]
    img_width  = int(round(img_height*26/36, 0))
    img_up     = int(round(img_height*19/36, 0))
    img_down   = int(round(img_height*25/36, 0))
    img_left   = int(round(img_width *11/26, 0))
    img_right  = int(round(img_width *23/26, 0))
    img = Image.fromarray(img[img_up:img_down, img_left:img_right])
    img.save(filename)
```

## Подготовка модели

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, BatchNormalization
from tensorflow.keras.optimizers import Adam
```

```python
imagedatagenerator = ImageDataGenerator

TRAINING_DIR = '/content/train'

train_datagen = imagedatagenerator(rescale=1.0 / 255.)

train_generator = train_datagen.flow_from_directory(TRAINING_DIR,
                                                    color_mode='grayscale',
                                                    batch_size=8,
                                                    class_mode='binary', # class_mode='categorical',
                                                    target_size=(64,64)
                                                    )

VALIDATION_DIR = '/content/test'

validation_datagen = imagedatagenerator(rescale=1.0 / 255.)

validation_generator = validation_datagen.flow_from_directory(VALIDATION_DIR,
                                                              color_mode='grayscale',
                                                              batch_size=8,
                                                              class_mode='binary', # class_mode='categorical',
                                                              target_size=(64,64)
                                                              )

model = Sequential()
model.add(BatchNormalization(input_shape=(64,64, 1)))
model.add(Conv2D(16, (3, 3), padding='same', activation='relu'))
model.add(Conv2D(16, (3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(32, (3, 3), padding='same', activation='relu'))
model.add(Conv2D(32, (3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
# model.add(Dense(1024, activation='relu'))
model.add(Dense(512, activation='relu'))
model.add(Dense(33, activation='softmax'))

model.compile(optimizer=Adam(lr=1e-3), 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
batch_size = 8

history = model.fit(train_generator,
                    steps_per_epoch=max(1, num_train//batch_size),
                    epochs=5,
                    verbose=1,
                    validation_data=validation_generator,
                    validation_steps=max(1, num_val//batch_size))

# history = model.fit_generator(train_generator,
#                               epochs=5,
#                               verbose=1,
#                               validation_data=validation_generator)

# model.save('/content/drive/My Drive/Projects/MindSet OCR/model.h5')
```
